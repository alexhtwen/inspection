````python
from tools import say
from operations import ArithmeticOperations, valid_operators

class Calculator():
    def __init__(self):
        ...

    def calculate(self, expression: str) -> int:
        return_value = None
        for valid_operator in valid_operators:
            if valid_operator in expression:
                partitioned_exp = expression.partition(valid_operator)
                operand1 = int(partitioned_exp[0])
                actual_operator = partitioned_exp[1]
                operand2 = int(partitioned_exp[2])
                operation = ArithmeticOperations()
                match actual_operator:
                    case '+':
                        return_value = operation.add(operand1, operand2)
                    case '-':
                        return_value = operation.subtract(operand1, operand2)
                    case '*':
                        return_value = operation.multiply(operand1, operand2)
                    case '/':
                        if operand2 == 0:
                            raise ValueError("Cannot divide by zero")
                        return_value = operation.divide(operand1, operand2)
                    case _:
                        raise ValueError("Invalid operator")
                break
        if return_value is None:
            raise ValueError("No operator found")
        return return_value

cal = Calculator()
say(cal.calculate(' 9 + 6 '), blank_lines=(0, 1))
say(cal.calculate(' 9 - 6 '), blank_lines=(0, 1))
say(cal.calculate(' 9 * 6 '), blank_lines=(0, 1))
say(cal.calculate(' 9 / 6 '), blank_lines=(0, 1))
````

````python
import inspect
import sys
from typing import Callable

class Calculator:
    def __init__(self):
        self.operations = self._discover_operations()

    def _discover_operations(self) -> dict[str, Callable[[int, int], int]]:
        """
        Dynamically discover and return a dictionary of supported operations.
        """
        operator_map = {
            'add': '+',
            'subtract': '-',
            'multiply': '*',
            'divide': '/',
        }

        current_module = sys.modules[__name__]
        operations = {}

        for name, func in inspect.getmembers(current_module, inspect.isfunction):
            say(name, func, blank_lines=0)
            if name in operator_map:
                operations[operator_map[name]] = func

        return operations

    def calculate(self, operator: str, operand1: int, operand2: int) -> int:
        """
        Perform the calculation based on the operator and operands.
        """
        if operator in self.operations:
            func = self.operations[operator]
            return func(operand1, operand2)
        else:
            raise ValueError(f"Unsupported operator: {operator}")

    # Define arithmetic operation functions
    def add(a: int, b: int) -> int:
        return a + b

    def subtract(a: int, b: int) -> int:
        return a - b

    def multiply(a: int, b: int) -> int:
        return a * b

    def divide(a: int, b: int) -> int:
        if b == 0:
            raise ValueError("Cannot divide by zero")
        return a / b

# Example usage
# if __name__ == "__main__":
# Create an instance of the Calculator
calculator = Calculator()

# Example inputs
operator = '+'
operand1 = 2
operand2 = 4

result = calculator.calculate(operator, operand1, operand2)
print(f"{operand1} {operator} {operand2} = {result}")
````

````python
import inspect
from typing import Callable, Any

class Calculator:
    def __init__(self):
        self.operations = self._discover_operations()

    def _discover_operations(self) -> dict[str, Callable[[int, int], int]]:
        """
        Dynamically discover and return a dictionary of supported operations.
        """
        operator_map = {
            'add': '+',
            'subtract': '-',
            'multiply': '*',
            'divide': '/',
        }

        operations = {}
        for name, func in inspect.getmembers(self, predicate=inspect.ismethod):
            if name in operator_map:
                sig = inspect.signature(func)
                doc = inspect.getdoc(func)
                operations[operator_map[name]] = {
                    'function': func,
                    'signature': sig,
                    'doc': doc
                }

        return operations

    def calculate(self, operator: str, operand1: int, operand2: int) -> int:
        """
        Perform the calculation based on the operator and operands.
        """
        if operator in self.operations:
            func_info = self.operations[operator]
            func = func_info['function']
            print(f"Using function: {func.__name__}")
            print(f"Signature: {func_info['signature']}")
            print(f"Documentation: {func_info['doc']}")
            return func(operand1, operand2)
        else:
            raise ValueError(f"Unsupported operator: {operator}")

    def add(self, operand1: int, operand2: int) -> int:
        """Add two numbers."""
        return operand1 + operand2

    def subtract(self, operand1: int, operand2: int) -> int:
        """Subtract second number from first."""
        return operand1 - operand2

    def multiply(self, operand1: int, operand2: int) -> int:
        """Multiply two numbers."""
        return operand1 * operand2

    def divide(self, operand1: int, operand2: int) -> int:
        """Divide first number by second. Raises ValueError if second number is zero."""
        if operand2 == 0:
            raise ValueError("Cannot divide by zero")
        return operand1 / operand2

# Example usage
if __name__ == "__main__":
    # Create an instance of the Calculator
    calculator = Calculator()

    # Example inputs
    operator = '+'
    operand1 = 10
    operand2 = 5

    result = calculator.calculate(operator, operand1, operand2)
    print(f"{operand1} {operator} {operand2} = {result}")
````
